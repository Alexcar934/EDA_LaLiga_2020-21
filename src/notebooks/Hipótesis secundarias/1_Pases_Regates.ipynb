# Pases y regates

En este notebook se obtienen:

    1. El número de pases acertados por mediocentro (ponderado por los minutos jugados)

    2. El número de regates acertados por mediocentro (ponderado por los minutoos jugados)
    

El objetivo es limpiar el dataset con el fin de dar respuesta a la hipótesis secundaria 1:

    "El número de pases acertados por un centrocampista es directamente proporcional al número de regates realizados por el mismo."

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/Users/alexcar934/Desktop/EDA_bien/src/data/raw/datasetMatchInfo2020-La-Liga.csv')
df = df.drop_duplicates()
```

```python
#Cogemos los mediocentros
df_cm = df[df['Pos'] == 'CM']
```

```python
'''Creamos un array donde aparezcan los nombres de todos los jugadores que, 
según el dataset, al menos una vez han jugado como centrocampista'''

df_cm_unique = df_cm['Player'].unique()
```

```python
'''
Creamos un bucle for que lea todos los nombres del array para 
que les asigne la posición CM en todos los partidos.
'''
for i in df_cm_unique:
    if np.any(df['Player'] == i):
        df['Pos'][df['Player']== i] = 'CM'
```

```python
df_cm = df[df['Pos'] == 'CM']
```

```python
#lo voy a medir por minutos más de 500, toques de balón totales, recuperaciones, asistencias, porcentaje de pases acertados.
df_minutos = df_cm.groupby('Player')['Min'].sum()
df_minutos.mean()
```

Seleccionamos únicamente los jugadores que hayan jugado más de 500 minutos

```python
df_minutos_cm = df_minutos.loc[df_minutos>500].sort_values(ascending =False)
df_minutos_cm.to_frame()
df_minutos_cm
```

Creamos el dataframe con la condición anterior

```python
df_cm= df_cm[df_cm['Player'].isin(df_minutos_cm.keys())]
```

Por otra parte, tendremos en cuenta únicamente los jugadores que hayan entrado en contacto con el balón 400 veces

```python

df_toques= df_cm.groupby('Player')['Performance|Touches'].sum()
df_toques_cm = df_toques.loc[df_toques>= (400)].sort_values(ascending=False)
df_toques_cm
```

Creamos el dataframe con la condición anterior

```python
df_cm= df_cm[df_cm['Player'].isin(df_toques_cm.keys())]
df_cm
```

Ponderamos los pases acertados entre los minutos que ha jugado cada uno.

```python
df_cmp_try= df_cm.groupby('Player')['Passes|Cmp'].sum().sort_values(ascending=False)
df_cmp= df_cmp_try[df_cmp_try.keys().isin(df_minutos_cm.keys())]
df_cmp = df_cmp*(df_minutos_cm/1440)
df_cmp = df_cmp.dropna().sort_values(ascending=False)
df_cmp
```

Ponderamos los regates acertados entre los minutos que ha jugado cada uno.

```python
df_drbls_try= df_cm.groupby('Player')['Dribbles|Succ_x'].sum().sort_values(ascending=False)
df_drbls_succ = df_drbls_try[df_drbls_try.keys().isin(df_minutos_cm.keys())]
df_drbls_succ = df_drbls_succ * df_minutos/1440
df_drbls_succ = df_drbls_succ.dropna().sort_values(ascending = False)
df_drbls_succ
```

Convertimos las series obtenidas en DataFrames

```python
df_cmp = df_cmp.to_frame()
df_drbls_succ = df_drbls_succ.to_frame()
```

Unimos los dataframes de pases y regates

```python

df_cm_final = df_drbls_succ.merge(df_cmp, how= 'inner', on=['Player'])
df_cm_final = df_cm_final.reset_index()
```

Cambiamos el nombre de las columnas y exportamos a csv

```python
df_cm_final.columns=['Jugador','Regates', 'Pases Acertados']
df_cm_final.to_csv('/Users/alexcar934/Desktop/EDA_bien/src/data/processed/csv_plots/1_Pases_Regates.csv')
df_cm_final

```
