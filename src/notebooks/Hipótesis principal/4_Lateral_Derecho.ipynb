# Lateral Derecho

Vamos a seleccionar al mejor lateral derecho según:

    1. Recuperaciones
    2. Porcentaje de pases acertados
    3. Asistencias
    4. Porcentaje de regates acerados 
    5. Goles encajados

Importamos librerías y el dataset inicial

```python
import pandas as pd
import numpy as np

#leemos el csv y eliminamos duplicidades

df = pd.read_csv('/Users/alexcar934/Desktop/EDA_bien/src/data/raw/datasetMatchInfo2020-La-Liga.csv')
df = df.drop_duplicates()
df

```

```python
'''Creamos un array donde aparezcan los nombres de todos los jugadores que, 
según el dataset, al menos una vez han jugado como lateral derecho'''
df_rb = df[df['Pos'] == 'RB']
df_rb_unique = df_rb['Player'].unique()
```

```python
'''
Creamos un bucle for que lea todos los nombres del array para 
que les asigne la posición RB en todos los partidos.
'''
for i in df_rb_unique:
    if np.any(df['Player'] == i):
        df.loc[(df['Player']== i), 'Pos'] = 'RB'

df[df['Player']=='Coke']
```

```python
'''
Para segmentar, establecemos que queremos obtener únicamente los partidos
en los que el futbolista haya jugado más de 45 minutos
'''
df_rb = df[df['Pos'] == 'RB']
df_rb =df_rb[df_rb['Min']>= 45]
df_rb
```

```python
#Creamos un dataframe con los minutos jugados por jugador
#Segmentamos por los jugadores que hayan jugado, al menos, un total de 720 minutos.
df_minutos = df_rb.groupby('Player')['Min'].sum()
df_minutos_rb = df_minutos.loc[df_minutos>720].sort_values(ascending =False)
df_minutos_rb
```

```python
#Creamos el nuevo dataframe con los futbolistas que han jugado más de 720 minutos
df_rb = df_rb[df_rb['Player'].isin(df_minutos_rb.keys())]
df_rb
```

```python

#como el mínimo de minutos es de 720 (más de 8 partidos), establezco un mínimo
#de 400 toques de balón en total 
df_toques= df_rb.groupby('Player')['Performance|Touches'].sum()
df_toques_rb = df_toques.loc[df_toques>= (50*8)].sort_values(ascending=False)
df_toques_rb
```

```python
#aplico el filtro al dataframe
df_rb = df_rb[df_rb['Player'].isin(df_toques_rb.keys())]
df_rb
```

```python
'''
Creamos un dataframe con la media de recuperaciones de balón por partido.
Posteriormente los ponderamos entre el total.
'''
df_recov_try= df_rb.groupby('Player')['Performance|Recov'].mean().sort_values(ascending=False)
df_recov = df_recov_try[df_recov_try.keys().isin(df_minutos_rb.keys())]
df_recov = df_recov/df_recov.sum()
df_recov
```

```python
'''
Creamos un dataframe con las asistencias realizadas por jugador.
'''
df_ast_try= df_rb.groupby('Player')['Performance|Ast'].sum().sort_values(ascending=False)
df_ast = df_ast_try[df_recov_try.keys().isin(df_minutos_rb.keys())]
df_ast 
```

```python
'''
Creamos un dataframe con la media de los porcentajes de pases acertados por partido.
Posteriormente los ponderamos entre el total.
'''
df_cmp_try= df_rb.groupby('Player')['Passes|Cmp%'].mean().sort_values(ascending=False)
df_cmp= df_cmp_try[df_cmp_try.keys().isin(df_minutos_rb.keys())]
df_cmp= df_cmp/df_cmp.sum()
df_cmp
```

```python
'''
Creamos un dataframe con la media de los porcentajes de regates acertados por partido.
Posteriormente los ponderamos entre el total.
'''
df_drbls_try= df_rb.groupby('Player')['Dribbles|Succ%'].sum().sort_values(ascending=False)
df_drbls_succ = df_drbls_try[df_drbls_try.keys().isin(df_minutos_rb.keys())]
df_drbls_succ = df_drbls_succ/(df_drbls_succ.sum())
df_drbls_succ
```

```python
#Convertimos las series a dataframe
df_recov=df_recov.to_frame()
df_cmp = df_cmp.to_frame()
df_ast = df_ast.to_frame()
```

```python
'''
Unimos todos los dataframes por la variable Player
'''
df_rb_final = df_recov.merge(df_cmp, how= 'inner', on=['Player'])
df_rb_final = df_rb_final.merge(df_ast, how='inner', on=['Player'])
df_rb_final = df_rb_final.merge(df_drbls_succ, how='inner', on=['Player'])
df_rb_final
```

```python
'''Leemos el csv de los porteros.
Esto se debe a que, en el csv que muestra a todos los jugadores y sus estadísticas,
no aparecen los goles encajados por cada jugador.
'''
df_gk = pd.read_csv('/Users/alexcar934/Desktop/EDA_bien/src/data/raw/datasetMatchGkInfo2020-La-Liga.csv')
df_gk = df_gk.drop_duplicates()
df_goalsallowed = df_gk.groupby(['Team','Match_x'])['Shot Stopping|GA'].mean()
```

```python
#añadimos la columna de goles encajados al dataframe general
df = df.merge(df_goalsallowed, how= 'inner', on=['Match_x', 'Team']).drop_duplicates()
```

```python
'''
Creamos un dataframe con la media de goles encajados por partido
y los minutos jugados. Segmentamos por los futbolistas que hayan jugado más de 720 minutos
'''
df_goals_try = df.groupby('Player')['Shot Stopping|GA'].mean()
df_goals_try = df_goals_try.to_frame()
df_goals_try['Min'] = df_minutos.sort_values()
df_goals_away=df_goals_try[df_goals_try['Min']>720]
df_goals_away
```

```python
'''Los goles encajados, a los defensas, los ponderaremos por los minutos que haya jugado cada uno'''
df_mins_percent = df_minutos_rb/1440
df_goals_mins = df_goals_away['Shot Stopping|GA']* df_mins_percent
df_goals_mins.sort_values(ascending = True)
```

```python
'''Ponderamos los goles encajados sobre el total'''
df_goals_mins= df_goals_mins.to_frame()
df_goals_mins = df_goals_mins/df_goals_mins.sum()
```

```python
'''Unimos el dataframe con los goles encajados y el de 
los equipos de cada jugador al dataframe final'''

df_rb_final = df_rb_final.merge(df_goals_mins, how='inner', on=['Player'])
df_rb_final = df_rb_final.reset_index()
df_teams= df[['Player','Team']]
df_rb_final=df_teams.merge(df_rb_final, how='inner', on='Player').drop_duplicates()
df_rb_final
```

```python
#establecemos la Puntuación Final

df_rb_final['Puntuación Final'] = (df_rb_final.iloc[:,2]) + (df_rb_final.iloc[:,3]) + (df_rb_final.iloc[:,4]*0.01) + (df_rb_final.iloc[:,5])-(df_rb_final.iloc[:,6])
df_rb_final= df_rb_final.sort_values(by='Puntuación Final', ascending=False) 
```

```python
#Cambiamos el nombre de las columnas
df_rb_final.columns = ['Jugador', 'Equipo', 'Recuperaciones', 'Porcentaje de pases acertados', 'Asistencias', 'Porcentaje de regates acertados', 'Goles encajados', 'Puntuación Final']
df_rb_final
```
