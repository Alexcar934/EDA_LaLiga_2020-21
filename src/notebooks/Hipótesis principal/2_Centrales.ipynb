# Centrales

Vamos a seleccionar a los 2 mejores defensas centrales según:

    1. Disparos Bloqueados.
    2. Intercepciones de balón.
    3. Recuperaciones 
    4. Despejes
    5. Goles encajados por partido

Importamos librerías y el dataset inicial

```python
'''Queremos obtener los dos mejores defensas del tramo de liga estudiado'''
import pandas as pd
import numpy as np

#leemos el csv y eliminamos duplicidades
df = pd.read_csv('/Users/alexcar934/Desktop/EDA_bien/src/data/raw/datasetMatchInfo2020-La-Liga.csv')
df = df.drop_duplicates()
df



```

Leemos el csv de los porteros.

Esto se debe a que, en el csv que muestra a todos los jugadores y sus estadísticas,
no aparecen los goles encajados por cada jugador.

```python
'''Leemos el csv de los porteros.
Esto se debe a que, en el csv que muestra a todos los jugadores y sus estadísticas,
no aparecen los goles encajados por cada jugador.
'''
df_gk = pd.read_csv('/Users/alexcar934/Desktop/EDA_bien/src/data/raw/datasetMatchGkInfo2020-La-Liga.csv')
df_gk = df_gk.drop_duplicates()
df_goalsallowed = df_gk.groupby(['Team','Match_x'])['Shot Stopping|GA'].sum()
df_goalsallowed
```

Unimos la columna de goles encajados al dataframe general

```python
#añadimos la columna de goles encajados al dataframe general
df = df.merge(df_goalsallowed, how= 'inner', on=['Match_x', 'Team']).drop_duplicates()

df 
```

Creamos un array donde aparezcan los nombres de todos los jugadores que, 
según el dataset, al menos una vez han jugado como defensas centrales'''

```python
df_cb = df[df['Pos'] == 'CB']
df_cb_unique = df_cb['Player'].unique()
df_cb_unique
```

```python
'''
Creamos un bucle for que lea todos los nombres del array para 
que les asigne la posición CB en todos los partidos.
'''
for i in df_cb_unique:
    if np.any(df['Player'] == i):
        df.loc[(df['Player']== i), 'Pos'] = 'CB'

df[df['Player']=='Coke']
```

```python
'''
Para segmentar, establecemos que queremos obtener únicamente los partidos
en los que el futbolista haya jugado más de 45 minutos
'''
df_cb = df[df['Pos'] == 'CB']
df_cb =df_cb[df_cb['Min']>= 45]
```

```python
#Creamos un dataframe con los minutos jugados por jugador
#Segmentamos por los jugadores que hayan jugado, al menos, un total de 900 minutos.

df_minutos = df_cb.groupby('Player')['Min'].sum()
df_minutos_cb = df_minutos.loc[df_minutos>=900].sort_values(ascending =False)
df_minutos_cb
```

```python
'''
Creamos un dataframe con la media de balones interceptados por partido.
Posteriormente los ponderamos entre el total
'''

df_int_try= df_cb.groupby('Player')['Performance|Int_x'].mean().sort_values(ascending=False)
df_int = df_int_try[df_int_try.keys().isin(df_minutos_cb.keys())]
df_int = df_int/df_int.sum()
```

```python
'''
Creamos un dataframe con la media de disparos bloqueados por partido.
Posteriormente los ponderamos entre el total.
'''
df_blocks_try= df_cb.groupby('Player')['Performance|Blocks'].mean().sort_values(ascending=False)
df_blocks = df_blocks_try[df_blocks_try.keys().isin(df_minutos_cb.keys())]
df_blocks = df_blocks/df_blocks.sum()
```

```python
'''
Creamos un dataframe con la media de recuperaciones de balón por partido.
Posteriormente los ponderamos entre el total.
'''
df_recov_try= df_cb.groupby('Player')['Performance|Recov'].mean().sort_values(ascending=False)
df_recov = df_recov_try[df_recov_try.keys().isin(df_minutos_cb.keys())]
df_recov = df_recov/df_recov.sum()
```

```python
'''
Creamos un dataframe con la media de despejes por partido.
Posteriormente los ponderamos entre el total.
'''
df_clr_try= df_cb.groupby('Player')['Clr'].mean().sort_values(ascending=False)
df_clr = df_clr_try[df_clr_try.keys().isin(df_minutos_cb.keys())]
df_clr = df_clr/df_clr.sum()
df_clr
```

```python
'''
Creamos un dataframe con la media de goles encajados por partido.
Posteriormente los ponderamos entre el total.
'''
df_goals_try = df_cb.groupby('Player')['Shot Stopping|GA'].mean()
df_goals_try = df_goals_try.to_frame()
df_goals_try['Min'] = df_minutos_cb.sort_values()
df_goals_away=df_goals_try[df_goals_try['Min']>900]

```

```python
'''Los goles encajados, a los defensas, los ponderaremos por los minutos que haya jugado cada uno'''
df_mins_percent = df_minutos_cb/1440
df_goals_mins = df_goals_away['Shot Stopping|GA']* df_mins_percent
df_goals_mins.sort_values(ascending = True)
```

```python
'''
Pasamos todas las series a dataframe
'''

df_clr.to_frame()
df_blocks = df_blocks.to_frame()
df_int = df_int.to_frame()
df_recov = df_recov.to_frame()
df_goals_mins=df_goals_mins.to_frame()
```

```python
'''
Unimos todos los dataframes por la variable Player
'''

df_cb_final = df_blocks.merge(df_int, how= 'inner', on=['Player'])
df_cb_final = df_cb_final.merge(df_recov, how='inner', on=['Player'])
df_cb_final = df_cb_final.merge(df_clr, how='inner', on=['Player'])
df_cb_final = df_cb_final.merge(df_goals_mins, how='inner', on=['Player'])


df_cb_final
 
```

```python
'''
Creamos tres columnas nuevas:
B+I+R+D = los valores ponderados sobre el total de los Disparos bloqueados, Intercepciones, Recuperaciones y Despejes
Goals IMP = los goles medios encajados por partido por cada jugador, ponderados sobre el total.
Puntuación Final = La diferencia de las dos columnas anteriores'''

df_cb_final['B+I+R+D'] = df_cb_final.iloc[:,0] + df_cb_final.iloc[:,1]+(df_cb_final.iloc[:,2]) +(df_cb_final.iloc[:,3])
df_cb_final['B+I+R+D'] = df_cb_final['B+I+R+D']/df_cb_final['B+I+R+D'].sum()
df_cb_final['GoalsIMP'] = (df_cb_final.iloc[:,4]/df_cb_final.iloc[:,4].sum())
df_cb_final['Puntuación Final'] = df_cb_final.loc[:,'B+I+R+D'] - df_cb_final.loc[:,'GoalsIMP']
df_defensas = df_cb_final.sort_values(by='Puntuación Final', ascending=False)
df_defensas
```

```python
#Cambiamos los nombres de las columnas
df_defensas.columns=['Disparos bloqueados', 'Intercepciones de balón','Recuperaciones', 'Despejes', 'Goles encajados por partido', 'D+I+R+D','Goles Ponderados','Puntuación final']
df_defensas
```

```python
#añadimos el equipo al que pertenece cada jugador al dataframe
df_teams= df[['Player','Team']]
df_defensas=df_teams.merge(df_defensas, how='inner', on='Player').drop_duplicates().sort_values(by='Puntuación final',ascending=False)
df_defensas
```
